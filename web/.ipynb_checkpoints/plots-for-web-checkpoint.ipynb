```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
x = np.arange(0,100,0.1)

freq = 0.1

y = np.sin(2*np.pi*freq*x)
plt.plot(x,y, c='blue')
```

```python
x = np.arange(0,100,0.1)
freq = 0.02
y = np.sin(2*np.pi*freq*x)
plt.plot(x,y, c='red')
```
